```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline 
```

```python
from pydataset import data
# "data" is a pandas DataFrame with IDs and descriptions.
# Call data() to see the entire list.
# To load a particular data set, enter its ID as an argument to data().
#titanic_data = data("Titanic")
# To see the information about a data set, give data() the dataset_id with -
#show_doc=True
#data("Titanic", show_doc=True)

notem = data('nottem')
notem.plot(x = 'time', y = 'nottem', title = 'Nottem line plot')
plt.ylabel('Temperature (F)')
plt.show()

print "We displayed the line plot here \nThe 'Nottem' data contains the average monthly temperatures in Nottingam, 1920-1939\
\n so we can see that in Nottingham there is definitely different seasons and the range of the temperature can really swing over the course of a year"

VAdeaths_data = data("VADeaths")
VAdeaths_data[['Rural Male', 'Rural Female', 'Urban Male', 'Urban Female']].plot(kind = 'bar')
plt.title('Deaths in Virginia')
plt.ylabel('number of deaths')
plt.show()
print "Here is a box graph of the Virginia deaths data \n\
this data is displaying all the deaths in Virginia over different 5 year buckets of time.\n\
The data is broken down by male and female as well as rural and urban to compare which groups\
have the most deaths. Also, you can see that the older you get there are more people that will pass away"

arbuthnot_data = data('Arbuthnot')
arbuthnot_data.plot(x = 'Year',y = 'Ratio')
plt.title('Ratio of Male Births to Female Births in London')
plt.ylabel('male births/female births')
plt.show()
print "This graph shows the ratio of male births to female births it is interesting to note that in about\
 every single year the ratio is over 1 meaning there were more male births than female so, they must have followed\
 the old legends of eating lots of pickles, steak, apples, avocado and broccoli."
```

# Problem 2

We examine certain datasets and plot histograms

```python
trees_data = data('trees')
road_data = data('road')
birthdeathrates_data = data('birthdeathrates')
print "The Trees data set contains Girth, Volume, and Height for Cherry Trees. We see that the Height is evenly distributed, but Volume and \
Girth have some outliers."
trees_data[['Girth',]].plot(kind = 'box')
plt.title('Boxplot of Girth of Cherry Trees')
plt.show()
trees_data[['Volume']].plot(kind = 'box')
plt.title('Boxplot of Volume Cherry Trees')
plt.show()
trees_data[['Height']].plot(kind = 'box')
plt.title('Boxplot of Height of Cherry Tree')
plt.ylabel('units')
plt.show()
#data('trees', show_doc = True)
print "Source: pydataset 'trees' \n This boxplot shows us that 75% of the black cherry trees have a height in between 70 and 80 inches.\n\
Also, it shows that the volume has a very wide range."
road_data[['deaths']].plot(kind = 'hist', label = True)
plt.title('Histogram of fatal car accidents')
plt.ylabel('Number of States')
plt.xlabel('deaths')
plt.show()
print "This histogram always us to see fatal death accidents based on a state level, meaning we can see how many states\
have 0-500, 500-1000 fatal accidents etc. we see that 17 of the 25 states have less than 1000 fatal accidents."
birthdeathrates_data[['birth','death']].plot(kind = 'hist', alpha = .5, bins = 26)
plt.ylabel("Number of Countries")
plt.xlabel('Birth/Death rate')
plt.title('Histogram of Countries death and birth rate')
plt.show()
print "This histogram shows us the number of countries that have certain death rates or birth rates. \n\
specifically a large majority of countries have a death rate below 15. While most countries have a birth rate about 20."
```
